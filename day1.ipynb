```python
import pandas as pd
import regex as re
with open('/Users/manaskaushik/Documents/AoC2022/day1.txt', 'r') as f:
    Lines = f.readlines()
print(Lines)
    
```

```python
atom=[]
atom1=[]    
for line in Lines:
    if not (re.findall("(?<!\d)\\n",line)):
        print(line)
        atom1.append(int(re.sub("\n","",line)))
        print(f"{atom1=}")
    else:
        atom.append(atom1)
        atom1=[]    
        print(f"{atom=}")        
```

```python
start=0
elvescount=0
newd=[]
for line in Lines:
    if not (re.findall("(?<!\d)\\n",line)):
        start+=int(re.sub("\n","",line))
    else:
        elvescount+=1
        print(f"{elvescount=} : {start=}")
        newd.append(start)
        start=0        
```

```python
df=pd.DataFrame(newd, columns=['food'])
df1=df.sort_values(by=['food'],ascending=False)
df2=df1.reset_index(drop=True)
df2
```

```python
atom=0
for x in range(3):
    atom+=df2.loc[x,'food']
print(atom)
```
