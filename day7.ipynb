```python
import regex as re

with open('/Users/manaskaushik/Documents/AoC2022/day7.txt', 'r') as f:
    Lines = f.readlines()

meta=[]
for index, x in enumerate(Lines):
    if "cd" in x:
        meta.append(index)

meta1=[]
for index, x in enumerate(meta):
    leta=[]
    if index==len(meta)-1:
        break
    else:
        leta.append(meta[index])
        leta.append(meta[index+1])
    meta1.append(leta)

meta2=[]    
for x in meta1:
    leta=[]
    leta=Lines[x[0]:x[1]]
    meta2.append(leta)    
```

```python
meta3=[]

for x in meta2:
    leta=[]
    for y in x:
        y=re.sub("\n","",y)
        leta.append(y)
    meta3.append(leta)
print(meta3)    
        
```

```python
dali={}
mali={}
for x in meta3:
    yeta=[]
    geta=0
    for y in x:
        if "cd" not in y:
            yeta.append(y)
            try:
                atom=re.search("\d+",y)[0]
                geta+=int(atom) 
                print(f"{geta=}")
            except:
                pass
        if "cd" in y:
            try:
                dali[re.search("cd (.+)",y).group(1)]=yeta
                mali[re.search("cd (.+)",y).group(1)]=geta
            except:
                pass
print(dali)
#print(f"{mali=}")    
```
