```python
import regex as re

with open(r'D:\Data\AoC2022\day7.txt', 'r') as f:
#with open(r'D:\Data\AoC2022\day7_test.txt', 'r') as f:
    Lines = f.readlines()

feta=[]
i=0
for index, x in enumerate(Lines):
#    Lines=re.sub("\n","",Lines)
    if "cd" in x:
        feta.append(Lines[i:index])
        i=index

print(feta)
#    for y in Lines:
    #print(Lines)
    
```

```python
meta=[]
dir=[]
for x in feta:
    leta=[]
    dleta=[]
    for y in x:
        if re.search("\$ cd (.+?)\\n",y):
            try:
                leta.append(re.search("\$ cd (.+?)\\n",y).group(1))        
                #identify root directory to create  structure    
                dleta.append(re.search("\$ cd (.+?)\\n",y).group(1))        
            except:
                pass
        if re.search("(\d+) \w",y):
            leta.append((re.search("(\d+) \w",y).group(1)))

        #create directory structure    
        if re.search("dir (\w+)\\n",y):
            if re.search("dir (\w+)\\n",y) != '..':
                dleta.append((re.search("dir (\w+)\\n",y).group(1)))
                print((re.search("dir (\w+)\\n",y).group(1)))                      

    meta.append(leta)
    dir.append(dleta)
    print(dleta)
print(dir)
```

```python
#transform directory structure to a dictionary with lists of subdirectories
from collections import defaultdict
dirst = defaultdict(list)
for x in dir:
    print(x)
    print(len(x))
    for z in range(len(x)):
        if x[0] not in ["..","/"]:
            dirst[x[0]].append(x[z])
print(dirst)
        
```

```python
#with a total size of at most 100000
geta=[]
greta={}
for x in meta:
    leta=[]
    size=0
    for y in x:
        if y.isnumeric():
            size+=int(y)
        else:
            leta.append(y)
    leta.append(size)
    if size<100000:
        geta.append(leta)
#print(geta)

for x in geta:
    print(x)
    if len(x)>1:
        if x[0] != "..": 
            greta[x[0]]=x[1]
print(greta)
```

```python
from collections import defaultdict

dirstsize = defaultdict(list)

print(dirst)
for x in dirst:
    print(dirst[x])
    size=0
    for z in dirst[x]:
        if len(dirst[x])>1:
            try:
                size+=int(greta[z])
            except:
                pass
    dirstsize[x]=size

print(dirstsize)
        
```
