```python
import regex as re
ruck=[]
ruck2=[]
with open('/Users/manaskaushik/Documents/AoC2022/day3.txt', 'r') as f:
    Lines = f.readlines()
    for y in Lines:
        Lines=re.sub("\n","",y)
        atom=(int(len(Lines)/2))
        comp1=set(Lines[0:atom])
        comp2=set(Lines[atom:])           
        if comp1.intersection(comp2):
            for x in comp1.intersection(comp2):
                ruck.append(x)
print(ruck)
```

```python
#Lowercase item types a through z have priorities 1 through 26.
#Uppercase item types A through Z have priorities 27 through 52.
import string
lowerl=list(string.ascii_lowercase)
upperl=list(string.ascii_uppercase)
alpha=lowerl + upperl
numer=[x for x in range(1,53)]
res = {}
for key in alpha:
    for value in numer:
        res[key] = value
        numer.remove(value)
        break
print(res)
```

```python
score=0
for x in ruck:
    score+=res.get(x)
print(score)
    
```
